```python
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(211)
plt.plot(t1, f(t1), 'go', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
# Creating three figure in which first one has two subplot

import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0.0, 2.0, 0.01)
s1 = np.sin(3*np.pi*t)
s2 = np.cos(4*np.pi*t)

plt.figure(1)
plt.subplot(211)
plt.plot(t, s1)
plt.subplot(212)
plt.plot(t, 2*s1,'g-')

plt.figure(2)
plt.plot(t, s2)

plt.figure(3)
plt.subplot(211)
plt.plot(t, s2, 'ro')


plt.show()
```

```python
# Working with text

The text() command can be used to add text in an arbitrary location, and 
the xlabel(), ylabel() and title() are used to add text in the indicated locations.
```

```python
import numpy as np
import matplotlib.pyplot as plt

a=np.arange(0,1.0,0.01)
plt.plot(a, a**2)

plt.title('GRAPH 1')
plt.text(0.2,0.8,'observation found', fontsize=20)

plt.xlabel('time')
plt.ylabel('distance')

plt.show()
```

```python
# Annotating text

import numpy as np
import matplotlib.pyplot as plt

ax = plt.subplot(111)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate('false positive', xy=(3, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05),)

plt.ylim(-2,2)
plt.show()
```

```python
# Invert Axes - You can use decreasing axes by flipping the normal order of the axis limits

import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0.0, 2.0, 0.01)
s1 = np.sin(3*np.pi*t)

plt.plot(t, s1)
plt.show()

plt.plot(t, s1)
plt.xlim(2,0)     # Inverting Axes 
plt.show()
plt.ylim(0.75,0.75)
plt.show()
```

# Histogram

Histogram is used to calculate the density

```python
# Simple Histogram

import numpy as np
from numpy import pi,sin,cos, tan
import matplotlib.pyplot as plt

x = np.random.randn(500)

plt.hist(x) #histogram

plt.show()
```

```python
# Simple Histogram - Adding parameters to hist() function

import numpy as np
from numpy import pi,sin,cos, tan
import matplotlib.pyplot as plt

x = np.random.randn(500)
y = 50

plt.hist(x, y,  density=True, facecolor='g', alpha=0.55)

plt.show()
```

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
```

# Pie-Chart:

```python
labels = 'Python','Java','R','C++'
sizes = [350,240,125,215]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']

explode = (0.1, 0, 0, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f %%', shadow=True)

plt.axis('equal')
plt.show()
```
